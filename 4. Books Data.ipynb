# Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
import requests
from bs4 import BeautifulSoup
```

```python
soup=requests.get('https://bookpage.com/reviews')
content=soup.content
```

```python
content
```

```python
soup=BeautifulSoup(soup.content)
soup
```

```python
soup.title.text
```

```python
Books=soup.find('div',{'class':"bp-well"})
```

```python
Books
```

```python
firstbook=Books.findAll('div',{'bp-block article-info'})
```

```python
Book_Name = []
Author_Name = []
Genre = []
Review = []

for first_book in firstbook:
    Book_Name.append(first_book.h4.text.replace('\n',""))
    Author_Name.append(first_book.p.text.replace('\n',''))
    Genre.append(first_book.find('p',{'class':'genre-links hidden-phone'}).find('a').text)
    Review.append(first_book.find('p',{'excerpt'}).text.replace('\n',''))
```

```python
Book_Name
```

```python
Author_Name
```

```python
Genre
```

```python
Review
```

```python
Books_Review=list(zip(Book_Name,Author_Name,Genre, Review))
```

```python
import pandas as pd
df=pd.DataFrame(Books_Review[:5],columns=['Book_Name','Author_Name','Genre', 'Review'])
```
