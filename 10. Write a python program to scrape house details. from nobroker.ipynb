## Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should include house title, location, area, emi and price

```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
url=requests.get('https://www.nobroker.in/flats-for-sale-in-mumbai_mumbai')
content=url.content
```

```python
content
```

```python
soup_data=soup(content)
```

```python
soup_data=soup(content,'html.parser')
print(soup_data)
```

```python
flats=soup_data.findAll('div',{'class':"nb__2JHKO"})
```

```python
title=[]
area = []
location= []

for i in flats:
    title.append(i.span.text)
    location.append(i.find('div',{'class':'nb__2CMjv'}).text)
    area.append(i.find('div',{'class':'nb__3oNyC'}).text)

    
```

```python
price = []
emi=[]
full_info = []

detail = soup_data.find_all('div',class_="font-semi-bold heading-6")
for i in detail:
    full_info.append(i.text)

#EMI
for i in range(1,len(full_info),3):
    emi.append(full_info[i])

# price
for i in range(2,len(full_info),3):
    price.append(full_info[i])
```

```python
House=list(zip(title,location,area,emi,price))
```
